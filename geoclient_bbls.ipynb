```python
def make_dir_if_not_exists(dir_path):
    outdir = dir_path
    if not path.exists(outdir):
        mkdir(outdir)
```

```python
make_dir_if_not_exists('/output')
```

```python
def fetch_and_read_bbls_data_into_df(url):
    try:
        r = requests.get(url)
        print('Status code', r.status_code)
        print('Content-type', r.headers['content-type'])
        print('Encoding', r.encoding)
        json =  r.json()
           
        return pd.DataFrame(json, columns = json[0])

    except Exception as e:
        print('Status code', r.status_code)
        print('Error reads: ', e)
        return e
```

```python
cached_bbls_data = {}
```

```python
bbls_dfs = []
for index, row in bbls_table.iterrows():
    b = row['borough']
    bl = row['block']
    l = row['lot']
    
    api_str = Template(external_data_sources['bbls']).safe_substitute({ 'b': b, 'bl': bl, 'l': l})
    if (api_str not in cached_bbls_data):        
        df = fetch_and_read_bbls_data_into_df(api_str)
        cached_bbls_data[api_str] = df
        
    bbls_dfs.append(df)
    
final_bbls_df = pd.concat(bbls_dfs)
```
