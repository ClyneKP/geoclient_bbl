#### External API URL

```python
external_data_sources = {
    # b, bl, and l should be borough, block, and lot codes
    'bbls': 'https://api.cityofnewyork.us/geoclient/v1/bbl.json?borough=$b&block=$bl&lot=$l&app_id=c1c8e876&app_key=57f1b1d1e1484706e2cb5ffa20d17988',
}
```

#### Import Relevant Packages

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
import os
from os import listdir, path, mkdir
from os.path import isfile, join
from string import Template
import numpy as np
```

#### Importing BBL List

```python
bbls_table = pd.read_csv('dof_missing_bbls.csv', skiprows=4, usecols=[3, 4, 5, 6], dtype=str, na_values=['NaN'])
```

```python
def make_dir_if_not_exists(dir_path):
    outdir = dir_path
    if not path.exists(outdir):
        mkdir(outdir)
```

```python
make_dir_if_not_exists('output')
```

```python
def fetch_and_read_bbls_data_into_df(url):
    try:
        r = requests.get(url)
        print('Status code', r.status_code)
        print('Content-type', r.headers['content-type'])
        print('Encoding', r.encoding)
        json =  r.json()
        
        
        return pd.DataFrame.from_dict(json_normalize(json), orient='columns' )

    except Exception as e:
        print('Status code', r.status_code)
        print('Error reads: ', e)
        return e
```

```python
cached_bbls_data = {}
```

```python
bbls_dfs = []
for index, row in bbls_table.iterrows():
    b = row['borough']
    bl = row['block']
    l = row['lot']
    
    api_str = Template(external_data_sources['bbls']).safe_substitute({ 'b': b, 'bl': bl, 'l': l})
    if (api_str not in cached_bbls_data):        
        df = fetch_and_read_bbls_data_into_df(api_str)
        cached_bbls_data[api_str] = df
        
    bbls_dfs.append(df)
```

```python
make_dir_if_not_exists('output/bbls')
```

```python
final_bbl_df = pd.concat(bbls_dfs)
final_bbl_df.to_csv('output/bbls/results.csv')
```
